```python
import pandas as pd
import numpy as np
import re
import functools
import operator
```

```python
df = pd.read_csv('data/lenta-ru-news.csv')
```

```python
df.columns
```

```python
clear_df = pd.DataFrame({})
clear_df['title'] = df.title.str.lower().str.replace(r"(\n|\t|\xa0)", ' ')
clear_df['text'] = df.text.str.lower().str.replace(r"(\n|\t|\xa0)", ' ').str.replace(r"\.\s+", r".")
clear_df['topic'] = df.topic.str.lower().str.replace(r"(\n|\t|\xa0)", ' ')
clear_df['tags'] = df.tags.str.lower().str.replace(r"(\n|\t|\xa0)", ' ')
```

```python
clear_df['text'] = clear_df['text'].str.split(".")
clear_df.dropna(inplace=True)
```

```python
all_sentences = functools.reduce(operator.iconcat, clear_df['text'], [])
```

```python
true_idx = [idx for idx, sent in enumerate(all_sentences) if sent is not '']
all_sentences = list(operator.itemgetter(*true_idx)(all_sentences))
```
