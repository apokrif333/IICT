```python
import pandas as pd
import numpy as np
import re
import functools
import operator
from tqdm import tqdm
```

```python
df = pd.read_csv('data/lenta-ru-news.csv')
```

```python
df.columns
```

```python
clear_df = pd.DataFrame({})
clear_df['text'] = df.text.str.lower().str.replace(r"(\n|\t|\xa0)", ' ').str.replace(r"\.\s+", r".").str.replace(
    r"(«|»|,|:|\(|\)|\[|\]|,)", "")
```

```python
clear_df['text'] = clear_df['text'].str.split(".")
clear_df.dropna(inplace=True)
```

```python
all_sentences = functools.reduce(operator.iconcat, clear_df['text'], [])
```

```python
true_idx = [idx for idx, sent in enumerate(all_sentences) if sent is not '']
all_sentences = list(operator.itemgetter(*true_idx)(all_sentences))
```

```python
all_sentences[0].split(' ')
```

```python
import tensorflow as tf
import tensorflow_hub as hub
elmo_tf = hub.Module("http://files.deeppavlov.ai/deeppavlov_data/elmo_ru-news_wmt11-16_1.5M_steps.tar.gz", trainable=True)
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
# embeddings = elmo_tf(["Я тяжело болен",
#                       "Я тяжело устал",
#                       "Вся тяжесть этого мира лежит на мне",
#                       "Это тяжёлый арбуз",
#                       "У меня тяжёлая жена",
#                       "У моего друга очень тяжёлый характер",
#                       "На мне было тяжёлое бремя прорицателя",
#                       "Тяжёлая доля - быть отцом взрослой дочери"
#                      ], 
#                      signature="default", as_dict=True)["elmo"]
# embeddings = elmo_tf(all_sentences, signature="default", as_dict=True)["lstm_outputs2"]
at = [all_sentences[0].split(), all_sentences[1].split()]
embeddings = elmo_tf(inputs={"tokens": [["я", "тяжёло", "устал"], ["арбуз", "тяжело", "нести"]],
                             "sequence_len": [3, 1]}, 
                     signature="tokens", as_dict=True)["elmo"]
print(sess.run(embeddings).shape)
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(sess.run(embeddings)[1])
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

pca.fit(x)
print(pca.explained_variance_ratio_,
     pca.singular_values_
     )
print(pca.fit_transform(x))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

for sentence_2d in sess.run(embeddings):
    pca.fit(sentence_2d)
    print(pca.explained_variance_ratio_,
          pca.singular_values_
         )
    print(pca.fit_transform(sentence_2d))
```

```python
part_sentenses = all_sentences[:100]

all_words = dict()
for sentence in all_sentences[:100]:
    new_keys = sentence.split(' ')
    all_words.update(
        dict(zip(new_keys, len(new_keys) * [[]]))
    )
all_words
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

for sentence in range(len(sess.run(embeddings))):
    pca_sentence = pca.fit_transform(sess.run(embeddings)[sentence])
    split_sentence = all_sentences[sentence].split(' ')
    for word in range(len(split_sentence)):
        print(split_sentence[word], pca_sentence[word])
        all_words[split_sentence[word]].append(pca_sentence[word])
    print('--------------------------------------')

all_words
```

```python
embeddings = sess.run(embeddings)
number_sentences = len(embeddings)
for sentence in tqdm(range(number_sentences)):
    split_sentence = all_sentences[sentence].split(' ')
    cur_sentence = embeddings[sentence]
    for word in range(len(split_sentence)):
        all_words[split_sentence[word]].append(cur_sentence[word])

all_words
```

```python
word = 'спецслужбами'
difference = []
for i in range(len(all_words[word])-1):
    difference.append(np.linalg.norm(all_words[word][i] - all_words[word][i+1]))
    
print(np.std(difference))
```
